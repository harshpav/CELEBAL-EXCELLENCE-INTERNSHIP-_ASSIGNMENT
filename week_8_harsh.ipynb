# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (IMPLEMENTED)

import logging

# Bonus: Add logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

def agent(query: str) -> dict:
    """
    Single-agent smart assistant with conditional routing.

    Routes:
      - Math queries (contains 'calculate') -> Calculator Tool
      - Keyword queries (contains 'keywords') -> Keyword Extractor Tool
      - General queries -> Direct response
    """
    logger.info(f"Received query: {query}")
    query_lower = query.lower()

    try:
        # Route 1: Calculator
        if "calculate" in query_lower:
            # Extract expression after the word 'calculate'
            expression = query_lower.replace("calculate", "").strip()
            logger.info(f"Routing to Calculator Tool with expression: {expression}")
            result = calculator(expression)
            return {
                "type": "calculation",
                "result": result
            }

        # Route 2: Keyword Extractor
        elif "keywords" in query_lower:
            # Extract text after 'from' if present, else use full query
            if "from" in query_lower:
                text = query[query_lower.index("from") + 4:].strip()
            else:
                text = query.strip()
            logger.info(f"Routing to Keyword Extractor Tool with text: {text}")
            result = extract_keywords(text)
            return {
                "type": "keywords",
                "result": result
            }

        # Route 3: General Response
        else:
            logger.info("Routing to General Response")
            return {
                "type": "general",
                "result": f'You asked: "{query}". This is a general query. Please use a language model for a detailed answer.'
            }

    except Exception as e:
        logger.error(f"Agent encountered an error: {e}")
        return {
            "type": "error",
            "result": f"An error occurred: {str(e)}"
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````
